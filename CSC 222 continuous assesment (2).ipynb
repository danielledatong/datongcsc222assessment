# Step 1: Import Libraries and Load the Dataset


```python
# Import necessary libraries
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\\Users\\HP\\Documents\\csc_222_student_records_(5000).csv")

# Inspect the dataset's basic structure
data_shape = data.shape
data_info = data.info()
data_summary = data.describe()
print(data_shape, data_info, data_summary)
```

# Step 2: Inspect the Dataset

```python
# Check for missing values
missing_values = data.isnull().sum()
print(missing_values)

```

# Step 3: Clean and Preprocess the Data


```python
# Fill missing values in selected columns with 'Not Given'
columns_to_fill = ['Certifications', 'Soft_Skills_Score', 'Networking_Score']
data[columns_to_fill] = data[columns_to_fill].fillna('Not Given')

# Recheck missing values after filling
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Drop irrelevant column (University_Ranking)
data_cleaned = data.drop(columns=['University_Ranking'])

# Confirm changes
data_cleaned_info = data.info()
```

# Step 5: Data Visualization


## 5.1 Visualization Using Matplotlib


```python
# Bar Chart: Number of Students by Field of Study
import matplotlib.pyplot as plt

field_counts = data['Field_of_Study'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(field_counts.index, field_counts.values, color='skyblue')
plt.title('Number of Students by Field of Study')
plt.xlabel('Field of Study')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Gender Distribution
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['red', 'blue', 'gold'])
plt.title('Gender Distribution')
plt.show()
```

```python
# Histogram: Distribution of University GPA
plt.figure(figsize=(8, 5))
plt.hist(data['University_GPA'].dropna(), bins=20, color='green', edgecolor='black')
plt.title('Distribution of University GPA')
plt.xlabel('GPA')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Line Chart: Average Starting Salary by Job Level
salary_by_level = data.groupby('Current_Job_Level')['Starting_Salary'].mean()
plt.figure(figsize=(8, 5))
plt.plot(salary_by_level.index, salary_by_level.values, marker='o', linestyle='--', color='purple')
plt.title('Average Starting Salary by Job Level')
plt.xlabel('Job Level')
plt.ylabel('Average Starting Salary')
plt.grid(True)
plt.show()
```

## 5.2 Visualization Using Seaborn



# Step 6: Brief Note on Seaborn and Its Importance in Data Analysis

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics with minimal coding effort. 

Unlike Matplotlib, which often requires a lot of manual styling, Seaborn comes with beautiful default styles and color palettes. It also works seamlessly with Pandas DataFrames, making it easy to plot data directly from tables without extra steps.

Seaborn is especially useful in data analysis bec1 se:
- It simplifies complex visualizations like boxplots, violin plots, heatmaps, and pair2lots.
- It helps reveal patterns, relationships, and correlations between multiple variables q3ickly.
- It produces polished, publication-quality graphics automa4ically.
- It integrates statistical summaries (like confidence intervals) into plots without em clearly.



```python
# Import seaborn for advanced visualizations
import seaborn as sns
sns.set(style='whitegrid')
```

```python
# Boxplot: University GPA by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='University_GPA', data=data, palette='pastel')
plt.title('University GPA by Gender')
plt.xlabel('Gender')
plt.ylabel('GPA')
plt.show()
```

```python
# Heatmap: Correlation between numerical features
plt.figure(figsize=(10, 8))
corr = data.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Scatterplot: GPA vs Networking Score by Gender
data['Networking_Score'] = pd.to_numeric(data['Networking_Score'], errors='coerce')
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Networking_Score', y='University_GPA', hue='Gender', data=data)
plt.title('GPA vs Networking Score by Gender')
plt.xlabel('Networking Score')
plt.ylabel('University GPA')
plt.show()
```

```python
# Pairplot: Relationships among Career-Related Scores
career_features = ['University_GPA', 'Soft_Skills_Score', 'Networking_Score', 'Career_Satisfaction']
sns.pairplot(data[career_features].dropna())
plt.suptitle('Pairplot of Career-Related Scores', y=1.02)
plt.show()
```
